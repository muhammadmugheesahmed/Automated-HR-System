```python
import pandas as pd
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python

# Load the CSV file with a space as the delimiter
data = pd.read_csv("/content/drive/MyDrive/chatbotquestions.csv",delimiter=",",names=["Questions","Answer"])

# Print the first few rows to verify
print(data.head(5))
```

```python
print(data.columns)
```

```python
Questions = data['Questions']
Answer = data['Answer']

print(Questions,Answer)
```

```python

# TF-IDF Vectorizer: Convert questions into TF-IDF vectors
vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(data['Questions'])  # Fit and transform all questions
```

```python

# Function to Find the Most Similar Question and Answer
def get_answer(user_question, data, tfidf_matrix, vectorizer):
    # Transform the user question into a TF-IDF vector
    user_tfidf = vectorizer.transform([user_question])

    # Calculate cosine similarity between user question and all questions
    cosine_similarities = cosine_similarity(user_tfidf, tfidf_matrix)

    # Find the index of the most similar question
    most_similar_idx = cosine_similarities.argmax()
    similarity_score = cosine_similarities[0, most_similar_idx]

    # Return the answer if similarity is above a threshold
    if similarity_score > 0.2:  # Threshold for relevance (can be tuned)
        return data['Answer'][most_similar_idx]
    else:
        return "Sorry, I don't have an answer for that."
```

```python
# Chatbot Interaction
print("Chatbot: Hello! Ask me a question or type 'exit' to stop.")
while True:
    user_input = input("You: ")
    if user_input.lower() == "exit":
        print("Chatbot: Goodbye!")
        break
    # Get the chatbot's response
    response = get_answer(user_input, data, tfidf_matrix, vectorizer)
    print("Chatbot:", response)
```
